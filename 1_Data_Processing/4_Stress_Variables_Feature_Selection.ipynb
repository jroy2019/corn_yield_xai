#### Adding Heat stress and Water stress markers at important plant growth stages 
* Heat stress, water stress  and water availabilty features are available in '6_Training_EC_Data_2014_2023'
* Environmental Covariates dataset is a curated dataset by Lopez et al. **https://www.nature.com/articles/s41467-023-42687-4**
* Environmental covariates generated by APSIM package using soil, weather and trait data

```python
#extract HI30 for plant growth stages between flower initation and end grain filling
#HI30 is number of days in given period where max temp > 30

hi30_columns = ['Env','HI30_pFloFla', 'HI30_pFlaFlw', 'HI30_pFlwStG','HI30_pStGEnG']
hi30_data = ec_data[hi30_columns]
```

```python
hi30_data.head(5)
```

```python
selected_ec_data = pd.concat([hi30_data, eo_data.drop(columns='Env')], axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler

def normalize_features(df: pd.DataFrame, exclude_cols: list = None):
    """
    Normalizes all numeric columns in a dataframe using MinMaxScaler, excluding any specified in exclude_cols. 
    Deletes the original columns after normalization.
    
    Returns:
        pd.DataFrame: DataFrame with normalized columns replacing original columns
    """
    df = df.copy()
    scaler = MinMaxScaler()
    
    if exclude_cols is None:
        exclude_cols = []

    for col in df.select_dtypes(include=['number']).columns: #only proceed if features are numeric regardless of whether it is int or float
        if col not in exclude_cols:
            df[col + '_Normalized'] = scaler.fit_transform(df[[col]])
            df.drop(columns=col, inplace=True)

    return df
```

```python
ec_data_normalised = normalize_features(selected_ec_data_with_water_levels, exclude_cols=['Env']) #do not normalise for merging with full dataset
```

#### Check missing year-locations in ec dataset

```python
#read in full, cleaned weather-snp-management dataset
print('importing large file ...')
full_dataset = pd.read_csv('final_boss_snp_weather_management.csv')
print('large file imported')
```

```python
ec_data_normalised = pd.read_csv('ec_data_normalised.csv')
```

```python
len(full_dataset['Env'].unique()), len(ec_data_normalised['Env'].unique())
#30+ missing year-locations in ec dataset
```

```python
import numpy as np

#create object with list of unique envs in soil dataset and full dataset
unique_ec_envs = ec_data_normalised['Env'].unique()
unique_full_envs = full_dataset['Env'].unique()

#find common envs in full and ec datasets
common_envs = np.intersect1d(unique_ec_envs, unique_full_envs)
```

```python
len(common_envs)
#240 year-locations in ec are in full dataset
#1 year-location in ec is not in full dataset (possibly removed after filtering)
```

```python
#find envs missing in ec data
missing_envs = np.setdiff1d(unique_full_envs, unique_ec_envs)

len(missing_envs)
#31 missing year-locations in ec dataset
```

```python
#convert to missing_envs to dataframe
missing_envs = pd.DataFrame(missing_envs)

missing_envs.to_csv('missing_envs_ec.csv', index=False)

#ec data for NCH1, IAH1 and ONH1, ONH1 and GEH1 (german) locations across most years are missing
#KSH1_2018 is also missing 
#IAH4_2014/2015 missing

#is imputation possible while maintaining fidelty. 
#same location different year = weather varies
#same year nearby location = soil could be varied
```
