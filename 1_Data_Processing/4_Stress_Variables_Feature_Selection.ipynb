```python
#extract HI30 for plant growth stages between flower initation and end grain filling
#HI30 is number of days in given period where max temp > 30

hi30_columns = ['Env','HI30_pFloFla', 'HI30_pFlaFlw', 'HI30_pFlwStG','HI30_pStGEnG']
df_hi30 = df_ec[hi30_columns]
```

```python
df_hi30.head(5)
```

```python
#Extract Eo for plant growth stages between emergence and end juvenile aswell as between flower initation and End grain fill
#Eo is Potential evaporation and transpiration of soil-plant system (water demand)

eo_columns = ['Env','Eo_pEmeEnJ', 'Eo_pFloFla', 'Eo_pFlaFlw', 'Eo_pFlwStG','Eo_pStGEnG']

df_eo = df_ec[eo_columns]
```

```python
df_eo.head(5)
```

```python
len(df_hi30), len(df_eo)
```

```python
df_ec_1 = pd.concat([df_hi30, df_eo.drop(columns='Env')], axis=1)
```

```python
df_ec_1.head(5)
```

```python
#the amount of water in layer 1-10 is available is Swmm_pGrowthStage_LayerNumber
#each layer is 20cm thick. Layer 1 is 0-20 cm, Layer 2 is 20-40 cm, and so on

swmm_columns = ['Env','SWmm_pEmeEnJ_1','SWmm_pEmeEnJ_2' , 'SWmm_pEmeEnJ_3','SWmm_pFloFla_4',
 'SWmm_pFlaFlw_4','SWmm_pFlwStG_4','SWmm_pStGEnG_4','SWmm_pFloFla_5',
 'SWmm_pFlaFlw_5','SWmm_pFlwStG_5','SWmm_pStGEnG_5','SWmm_pFloFla_6',
 'SWmm_pFlaFlw_6','SWmm_pFlwStG_6','SWmm_pStGEnG_6']

#water availiabilty of layer 1,2,3 for growth stages between emergence and end juvenile (0 - 60 cm)
#water availiabilty of layer 4,5,6 for growth stages between flowering and end grain fill (60-120 cm)


df_swmm = df_ec[swmm_columns]
```

```python
df_swmm.head(5)
```

```python
df_ec_2 = pd.concat([df_ec_1, df_swmm.drop(columns='Env')], axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler

def normalize_features(df: pd.DataFrame, exclude_cols: list = None):
    """
    Normalizes all numeric columns in a dataframe using MinMaxScaler, excluding any specified in exclude_cols. 
    Deletes the original columns after normalization.
    
    Returns:
        pd.DataFrame: DataFrame with normalized columns replacing original columns
    """
    df = df.copy()
    scaler = MinMaxScaler()
    
    if exclude_cols is None:
        exclude_cols = []

    for col in df.select_dtypes(include=['number']).columns: 
        if col not in exclude_cols:
            df[col + '_Normalized'] = scaler.fit_transform(df[[col]])
            df.drop(columns=col, inplace=True)

    return df

df_normalised = normalize_features(df_ec_2, exclude_cols=['Env'])
```
